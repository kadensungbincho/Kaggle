```python
col_list = result_cor_percent_low_rmtarget

def col_lin_check(row_name, df, result_list, check_num=0.6):
    for col in col_list:
        if row_name == col:
            break
        elif df.loc[row_name][col] >= check_num:
            result_list.append(row_name)
            return result_list
        
result_lst = []
for row in result_cor_percent_low_rmtarget.index:
    col_lin_check(row, result_cor_percent_low, result_lst)
    
set(result_lst)
print(len(result_lst))
print(result_lst)
```
