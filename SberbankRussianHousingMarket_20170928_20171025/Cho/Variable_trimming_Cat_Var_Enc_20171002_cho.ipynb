```python
train = pd.read_csv('dss-regression-datasets/project-sberbank-housing-market-data/train.csv')
train.head()
```

```python
cat_var_df = train.select_dtypes(include=['object']).copy()
del cat_var_df['timestamp']
cat_var_df.tail(n=30)
```

```python
cat_var_df[cat_var_df.isnull().any(axis=1)]
```

```python
cat_var_unique = []
for v in cat_var_df.columns:
    temp_uni_value = list(set(cat_var_df[v]))
    cat_var_unique.append(temp_uni_value) 

cat_var_unique 
```

```python
set(cat_var_df['sub_area'].values)
```

```python
cat_var_enc_sub_area = cat_var_df['sub_area'].value_counts()
cat_var_enc_sub_area.index
len(cat_var_enc_sub_area)
sub_area_enc_dict = {}
i = 1
for key in cat_var_enc_sub_area.index:
    sub_area_enc_dict[key] = i
    i += 1
sub_area_enc_dict
```

```python
cat_var_df.columns
```

```python
cat_var_yn = ['culture_objects_top_25', \
       'thermal_power_plant_raion', 'incineration_raion', \
       'oil_chemistry_raion', 'radiation_raion', 'railroad_terminal_raion', \
       'big_market_raion', 'nuclear_reactor_raion', 'detention_facility_raion', \
       'water_1line', 'big_road1_1line', 'railroad_1line']
type(cat_var_yn[0])
```

```python
cleanup_nums = {"product_type":{'OwnerOccupier' : 0, 'Investment' : 1},                
               "ecology": {'no data' : 0, 'poor' : 1, 'satisfactory' : 2, 'good' : 3, 'excellent' : 4}}

for v in cat_var_yn:
    cleanup_nums[v] = {'no':0, 'yes':1}
    
cleanup_nums['sub_area'] = sub_area_enc_dict

cleanup_nums
             
```

```python
cat_var_df.replace(cleanup_nums, inplace=True)
```
